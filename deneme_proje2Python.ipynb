```python
#IMDB Puanı Yuksek ilk 10 Film Bilgisi
IMDB_Score = dataRead["IMDB Score"]
IMDB_Title = dataRead["Title"]



filmScore = []
for degerPuan in IMDB_Score:
  filmScore.append(degerPuan)
print(filmScore)

filmTitle = []
for degerTitle in IMDB_Title:
  filmTitle.append(degerTitle)
print(filmTitle)


enBuyukSayi = 0


for islemSayi in range(10):
  enBuyukSayi = max(filmScore)
  print("\nEn Buyuk Degere Sahip IMDB Degeri:     ",enBuyukSayi)
  degerSirasiIMDB = filmScore.index(enBuyukSayi)
  print("En Cok IMDB Puanına Sahip Filmin Ismi:   ", filmTitle[degerSirasiIMDB])
  print("\n")
  filmScore.remove(enBuyukSayi)
```
